```python
# 25. Класс – отрезок. Методы – расчет середины и длины. Поля класса – координаты начала, конца, середины и длина.

import math

class Segment:
    def __init__(self, start: tuple[float], end: tuple[float]):
        if len(start) != len(end):
            raise ValueError('Разные размерности')
        self.__start = start
        self.__end = end

    @property
    def start(self):
        return self.__start

    @property
    def end(self):
        return self.__end

    @property
    def center(self):
        res = []
        for i in range(len(self.__start)):
            res.append((self.__end[i] - self.__start[i]) / 2)
        return tuple(res)

    @property
    def length(self):
        res = 0
        for i in range(len(self.__start)):
            res += (self.__end[i] - self.__start[i]) ^ 2
        res = math.sqrt(res)
        return res
```

```python
s = Segment((1, 2), (5, 7))
print(s.start)
print(s.end)
print(s.center)
print(s.length)
```

```python
# 26. Опишите класс Angle, заданный величиной в градусах и минутах (двумя полями). Включите в описание класса методы,
# позволяющие реализовать перевод в радианы, привести величину угла к диапазону 0–360°,
# увеличить и уменьшить угол на заданную величину и свойство, позволяющее определить, является ли угол тупым.

import math

class Angle:
    def __init__(self, degrees, minutes):
        self.__degrees = degrees
        self.__minutes = minutes

    @property
    def radians(self):
        d, m = self.__get_small_values()
        return (d + m / 60) / 180 * math.pi

    def convert_range(self):
        self.__degrees, self.__minutes = self.__get_small_values()

    def __get_small_values(self):
        d = self.__degrees
        m = self.__minutes
        if d < 0:
            d += 360


        d += ((math.copysign(1, self.__minutes) * self.__minutes) // 60) * math.copysign(1, self.__degrees)
        m -= 60 * ((math.copysign(1, self.__minutes) * self.__minutes) // 60) * math.copysign(1, self.__minutes)
        # print(d, ((math.copysign(1, self.__minutes) * self.__minutes) // 60), math.copysign(1, self.__degrees))
        d -= 360 * ((math.copysign(1, self.__degrees) * self.__degrees) // 360) * math.copysign(1, self.__degrees)

        return d, int(m)

    @property
    def degrees(self):
        return self.__degrees

    @property
    def minutes(self):
        return self.__minutes

    def add(self, degrees, minutes):
        self.__degrees += degrees
        self.__minutes += minutes

        if self.__minutes >= 60:
            self.__degrees += (math.copysign(1, self.__minutes) * self.__minutes) // 60
            self.__minutes -= 60 * ((math.copysign(1, self.__minutes) * self.__minutes) // 60)

    def subtract(self, degrees, minutes):
        self.add(-degrees, -minutes)

    @property
    def is_blunt(self):
        total = self.__degrees + self.__minutes / 60
        if total >= 360:
            total -= 360 * (int(total) // 360)
        return int(total) > 90
```

```python
a = Angle(80, 0)
print(f'{a.degrees} {a.minutes} {a.radians}')
print(a.is_blunt)
a.add(400, 5)
print(f'{a.degrees} {a.minutes} {a.radians}')
print(a.is_blunt)
a.convert_range()
print(f'{a.degrees} {a.minutes} {a.radians}')
a.subtract(400, 5)
print(f'{a.degrees} {a.minutes} {a.radians}')
print(a.is_blunt)
a.convert_range()
print(f'{a.degrees} {a.minutes} {a.radians}')
```

```python
# 27. Создайте класс ТОВАР с методами, позволяющими вывести на экран информацию о товаре, а также определить,
# предназначен ли он для заданного возраста потребителя. Создайте дочерние классы ИГРУШКА (название, цена,
# производитель, материал, возраст, на который рассчитана), КНИГА (название, автор, цена, издательство,
# возраст, на который рассчитана), СПОРТИНВЕНТАРЬ (название, цена, производитель, возраст, на который рассчитан)
# со своими методами вывода информации на экран и определения соответствия возрасту потребителя.
# Создайте список из  n  товаров, выведите полную информацию из базы на экран, а также организуйте поиск товаров
# для потребителя в заданном возрастном диапазоне.

from typing import Sequence

class Product:
    age: Sequence[int] = None

    @property
    def info(self):
        return None

    def __str__(self):
        return self.info

    def _ageToStr(self):
        return f'[{self.age[0] if self.age[0] != -1 else 0} - {self.age[1] if self.age[1] != -1 else "не ограничено"}]'

    def suitableForAge(self, age):
        if not age:
            raise NotImplemented()
        return self.age[0] < age < self.age[1] or \
            self.age[1] > age and self.age[0] == -1 or \
            self.age[0] < age and self.age[1] == -1 or \
            self.age[0] == -1 and self.age[1] == -1

class Toy(Product):
    def __init__(self, name, price, manufacturer, material, age: Sequence[int]):
        if len(age) != 2:
            raise ValueError('Аргумент age должен быть диапазоном из 2 чисел')
        self.name = name
        self.price = price
        self.manufacturer = manufacturer
        self.material = material
        self.age = age

    @property
    def info(self):
        return f'{self.name}, цена: {self.price}, производитель: {self.manufacturer}, материал: {self.material}, ' \
               f'возраст: {self._ageToStr()}'

class Book(Product):
    def __init__(self, name, price, author, publishing_office, age: Sequence[int]):
        if len(age) != 2:
            raise ValueError('Аргумент age должен быть диапазоном из 2 чисел')
        self.name = name
        self.price = price
        self.author = author
        self.publishing_office = publishing_office
        self.age = age

    @property
    def info(self):
        return f'{self.name}, цена: {self.price}, издательство: {self.publishing_office}, ' \
               f'возраст: {self._ageToStr()}'

class SportsEquipment(Product):
    def __init__(self, name, price, manufacturer, age: Sequence[int]):
        if len(age) != 2:
            raise ValueError('Аргумент age должен быть диапазоном из 2 чисел')
        self.name = name
        self.price = price
        self.manufacturer = manufacturer
        self.age = age

    @property
    def info(self):
        return f'{self.name}, цена: {self.price}, производитель: {self.manufacturer}, ' \
               f'возраст: {self._ageToStr()}'

class Storage:
    __l: list[Product]
    
    def __init__(self):
        self.__l = []

    def add(self, product: Product):
        self.__l.append(product)

    def print_all(self):
        print('Полный список:')
        for pr in self.__l:
            print(pr)

    def print_all_for_age(self, age):
        print(f'Подходит для {age} лет:')
        for pr in self.__l:
            if pr.suitableForAge(age):
                print(pr)
```

```python
storage = Storage()
storage.add(Toy(name='Кубик-рубик', price=400, manufacturer='Algeht', material='пластик', age=(6, 1)))
storage.add(Toy(name='Мяч', price=150, manufacturer='Litaft', material='резина', age=(-1, -1)))
storage.add(Book(name='Война и мир', price=150, author='Л.Н. Толстой', publishing_office='Издательство "Читаем"', age=(10, -1)))
storage.add(Book(name='Сказки', price=150, author='Л.Н. Толстой', publishing_office='"21 Век"', age=(-1, 10)))
storage.add(SportsEquipment(name='Гантеля', price=1500, manufacturer='Atucn Inc.', age=(16, -1)))
storage.add(SportsEquipment(name='Гиря 5 кг', price=500, manufacturer='Atucn Inc.', age=(10, -1)))
storage.add(SportsEquipment(name='Гиря 10 кг', price=650, manufacturer='Atucn Inc.', age=(15, -1)))

storage.print_all()
print()
storage.print_all_for_age(3)
print()
storage.print_all_for_age(12)
print()
storage.print_all_for_age(20)
```
